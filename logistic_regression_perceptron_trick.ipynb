```python
X, y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0, n_clusters_per_class=1, n_classes=2, random_state=41, hypercube=False, class_sep=10)
```

### Perceptron code from scratch

```python
class LogisticRegression():
  def __init__(self, learning_rate=0.1, max_iter=100):
    self.learning_rate = learning_rate
    self.max_iter = max_iter
    self.dW = None
    self.db = None
```

```python
def perceptron(self, X, y):
  X = np.insert(X, 0, 1, axis=1)
  weights = np.ones(X.shape[1])

  for i in range(self.max_iter):
    j = np.random.randint(0, X.shape[0] - 1)
    y_ = self.step(X[j], weights)
    weights = weights + self.learning_rate * (y[j] - y_) * X[j]

  self.dW = weights[1:,]
  self.db = weights[0]
  self.coef_ = self.dW
  self.intercept_ = self.db
  return self

LogisticRegression.perceptron = perceptron
```

```python
def step(self, X, weights):
  y_ = np.dot(X, weights)
  if (y_ >= 0): return 1
  return 0

LogisticRegression.step = step
```

```python
lr = LogisticRegression(learning_rate=0.1, max_iter=1000)
```

```python
lr.perceptron(X, y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
m = -(lr.coef_[0]/lr.coef_[1])
b = -(lr.intercept_/lr.coef_[1])
```

```python
x_input = np.linspace(-3, 3, X.shape[0])
y_input = m*x_input + b
```

```python
plt.plot(x_input, y_input, color='red', linewidth=2)
plt.scatter(X[:,0], X[:,1], c=y, cmap='winter')
plt.ylim(-3, 2)
plt.show()
```
