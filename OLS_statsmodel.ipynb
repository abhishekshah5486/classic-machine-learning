```python
df = pd.read_csv('cars24-car-price-clean.csv')
df.head()
```

```python
import statsmodels.api as sm
```

```python
y = df['selling_price']
X = df.drop(columns=['selling_price'], axis=1)
```

```python
model = sm.OLS(y, X)
```

```python
results = model.fit()
```

```python
results.summary()
```

```python
X_sm = sm.add_constant(X)
```

```python
X_sm.head()
```

```python
sm_model = sm.OLS(y, X_sm)
```

```python
sm_results = sm_model.fit()
```

```python
sm_results.summary()
```

### Multicollinearity: Variance Inflation Factor (VIF)

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
X_t = X
vif['Features'] = X_t.columns
vif['VIF_Values'] = [variance_inflation_factor(X_t.values, i) for i in range(X_t.shape[1])]
vif.sort_values(by='VIF_Values', ascending=False)
```

```python
cols2 = vif['Features'][1:,].values
cols2
```

```python
X2 = X[cols2]
```

```python
vif = pd.DataFrame()
X_t = X2
vif['Features'] = cols2
vif['VIF_Values'] = [variance_inflation_factor(X_t.values, i) for i in range(X_t.shape[1])]
vif = vif.sort_values(by='VIF_Values', ascending=False)

cols2 = vif['Features'][1:,].values
X2 = X[cols2]
```

```python
vif = pd.DataFrame()
X_t = X2
vif['Features'] = cols2
vif['VIF_Values'] = [variance_inflation_factor(X_t.values, i) for i in range(X_t.shape[1])]
vif = vif.sort_values(by='VIF_Values', ascending=False)

cols2 = vif['Features'][1:,].values
X2 = X[cols2]
cols2
```

```python
X2_sm = sm.add_constant(X2)
X2_sm_model = sm.OLS(y, X2_sm)
X2_sm_details = X2_sm_model.fit()
```
