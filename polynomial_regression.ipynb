```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import random
```

```python
X = 6 * np.random.rand(200, 1) - 3
y = 0.8 * X**2 + 0.9 * X + 2 + np.random.randn(200, 1)

# y = 0.8X**2 + 0.9X + 2 ( Polynomial of degree 2 )
```

```python
plt.scatter(X, y, color='blue')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
y_pred = lr.predict(X_test)
```

```python
plt.plot(X_train, lr.predict(X_train), color='red')
plt.scatter(X_train, y_train, color='blue')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polynomial = PolynomialFeatures(degree = 2)
```

```python
X_train_trans = polynomial.fit_transform(X_train)
X_test_trans = polynomial.fit_transform(X_test)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train_trans, y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.scatter(X_train, lr.predict(X_train_trans), color='red')
plt.scatter(X_train, y_train, color='blue')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
r2_score(y_test, lr.predict(X_test_trans))
```

### Multiple Polynomial Regression

```python
X = 6 * np.random.rand(200, 1) - 3
y = 6 * np.random.rand(200, 1) - 3
z = X**2 + y**2 + 0.5*y + 0.1*X*y + 0.9*X + 2 + np.random.randn(200, 1)

# z = X**2 + y**2 + 0.5*y + 0.1*X*y + 0.9*X + 2( Polynomial of degree 2 )
```

```python
n_X = pd.DataFrame({
    'X': X.flatten(),
    'y': y.flatten(),
})
```

```python
X_train, X_test, z_train, z_test = train_test_split(n_X, z, test_size=0.2, random_state=12)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, z_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(z_test, lr.predict(X_test))
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polynomial = PolynomialFeatures(degree = 2)
X_train_trans = polynomial.fit_transform(X_train)
X_test_trans = polynomial.fit_transform(X_test)
```

```python
X_train_trans
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train_trans, z_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(z_test, lr.predict(X_test_trans))
```
